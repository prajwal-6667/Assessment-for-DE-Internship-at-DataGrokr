```python
from sqlalchemy import create_engine, text
import pandas as pd

# Defining the SQLite database file path
db_path = 'house_database.db'

# Creating a SQLAlchemy engine to connect to the SQLite database
engine = create_engine(f'sqlite:///{db_path}')

# Defining a function to print the first few rows of a table
def print_table(table_name):
    query = f"SELECT * FROM {table_name} LIMIT 5"
    with engine.connect() as conn:
        result = conn.execute(text(query))
        df = pd.DataFrame(result.fetchall(), columns=result.keys())
    print(f"Table: {table_name}")
    print(df)

# Printing the first few rows of each table
tables_to_print = ['Property_Details', 'Property_Price_Details']
for table in tables_to_print:
    print_table(table)
```

# 1.Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1).

```python
# Query 1: Retrieve properties with a price greater than 1 million in "Estados Unidos" (l1)
query1 = """
SELECT PD.*
FROM Property_Details PD
JOIN Property_Price_Details PPD ON PD.id = PPD.id
WHERE PPD.price > 1000000 AND PD.l1 = 'Estados Unidos';
"""

# Executing the query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results1 = connection.execute(text(query1))
    df1 = pd.DataFrame(results1.fetchall(), columns=results1.keys())

# Printing the resulting DataFrame
print("\nQuery 1 Results:")
print(df1)
```

# 2.Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 'Medium' if it's between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters:

```python
# Query 2: Categorize properties based on their surface area
# 'Small' if less than 50 square meters, 'Medium' if between 50 and 100 square meters, 'Large' if greater than 100 square meters
query2 = """
SELECT PD.*,
    CASE
        WHEN PD.surface_total < 50 THEN 'Small'
        WHEN PD.surface_total >= 50 AND PD.surface_total <= 100 THEN 'Medium'
        ELSE 'Large'
    END AS surface_category
FROM Property_Details PD;
"""

# Executing the query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results2 = connection.execute(text(query2))
    df2 = pd.DataFrame(results2.fetchall(), columns=results2.keys())

# Printing the resulting DataFrame
print("\nQuery 2 Results:")
print(df2)
```

# 3.List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and bathrooms as another property in the dataset:

```python
# Query 3: List properties in "Belgrano" (l3) with the same number of bedrooms and bathrooms as another property
query3 = """
SELECT PD.id
FROM Property_Details PD
WHERE PD.l3 = 'Belgrano' AND (PD.bedrooms, PD.bathrooms) IN (
    SELECT PD2.bedrooms, PD2.bathrooms
    FROM Property_Details PD2
    GROUP BY PD2.bedrooms, PD2.bathrooms
    HAVING COUNT(*) > 1
);
"""

# Executing the query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results3 = connection.execute(text(query3))
    df3 = pd.DataFrame(results3.fetchall(), columns=results3.keys())

# Printing the resulting DataFrame
print("\nQuery 3 Results:")
print(df3)
```

# 4.Calculate the average price per square meter (price / surface_total) for each property type (property_type) in the "Belgrano" neighborhood (l3):

```python
# Query 4: Calculate average price per square meter for each property type (property_type) in "Belgrano" (l3)
query4 = """
SELECT property_type, AVG(price / surface_total) AS avg_price_per_sqm
FROM Property_Price_Details,Property_Details
WHERE l3 = 'Belgrano'
GROUP BY property_type;
"""

# Executing the query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results4 = connection.execute(text(query4))
    df4 = pd.DataFrame(results4.fetchall(), columns=results4.keys())

# Printing the resulting DataFrame
print("\nQuery 4 Results:")
print(df4)
```

# 5.Identify properties that have a higher price than the average price of properties with the same number of bedrooms and bathrooms.

```python
# Query 5: Identify properties that have a higher price than the average price of properties with the same number of bedrooms and bathrooms
query5 = """
SELECT pd.id, pd.bedrooms, pd.bathrooms, pp.price
FROM Property_Details pd
INNER JOIN Property_Price_Details pp ON pd.id = pp.id
WHERE pp.price > (
    SELECT AVG(price)
    FROM Property_Price_Details
    WHERE bedrooms = pd.bedrooms
    AND bathrooms = pd.bathrooms
) LIMIT 10;
"""

# Executing the corrected query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results5 = connection.execute(text(query5))
    df5 = pd.DataFrame(results5.fetchall(), columns=results5.keys())

# Printing the resulting DataFrame
print("\nCorrected Query 5 Results:")
print(df5)
```

# 6.Calculate the cumulative price for each property type, ordered by the creation date.

```python
# Query 6: Calculate the cumulative price for each property type, ordered by the creation date.
query6 = """
SELECT PD.l3, SUM(PPD.price) AS cumulative_price
FROM Property_Details PD
JOIN Property_Price_Details PPD ON PD.id = PPD.id
GROUP BY PD.l3
ORDER BY cumulative_price DESC
LIMIT 10;
"""

# Executing the corrected query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results6 = connection.execute(text(query6))
    df6 = pd.DataFrame(results6.fetchall(), columns=results6.keys())

# Printing the resulting DataFrame
print("\nCorrected Query 6 Results:")
print(df6)
```

# 7.Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for sale (operation_type = 'Venta')

```python
# Query 7: Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for sale (operation_type = 'Venta')
query7 = """
SELECT PD.l3, SUM(PD.surface_total) AS total_surface_area
FROM Property_Details PD
GROUP BY PD.l3
ORDER BY total_surface_area DESC
LIMIT 10;
"""

# Executing the corrected query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results7 = connection.execute(text(query7))
    df7 = pd.DataFrame(results7.fetchall(), columns=results7.keys())

# Printing the resulting DataFrame
print("\nCorrected Query 7 Results:")
print(df7)
```

# 8.Find the top 5 most expensive properties in "Palermo" (l3) listed in August 2020

```python
# Query 8: Find the top 5 most expensive properties in "Palermo" (l3) listed in August 2020
query8 = """
SELECT PD.*
FROM Property_Details PD
JOIN Property_Price_Details PPD ON PD.id = PPD.id
WHERE PD.l3 = 'Palermo' AND strftime('%Y-%m', PD.created_on) = '2020-08'
ORDER BY PPD.price DESC
LIMIT 5;
"""

# Executing the query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results8 = connection.execute(text(query8))
    df8 = pd.DataFrame(results8.fetchall(), columns=results8.keys())

# Printing the resulting DataFrame
print("\nQuery 8 Results:")
print(df8)
```

# 9.Find the top 3 properties with the highest price per square meter within each property type.

```python
# Query 9: Find the top 3 properties with the highest price per square meter within each property type.
query9 = """
WITH RankedProperties AS (
    SELECT PD.*, PPD.price / PD.surface_total AS price_per_sqm,
        ROW_NUMBER() OVER (PARTITION BY PD.l1, PD.l2, PD.l3 ORDER BY PPD.price / PD.surface_total DESC) AS rank
    FROM Property_Details PD
    JOIN Property_Price_Details PPD ON PD.id = PPD.id
)
SELECT *
FROM RankedProperties
WHERE rank <= 3;
"""

# Executing the corrected query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results9 = connection.execute(text(query9))
    df9 = pd.DataFrame(results9.fetchall(), columns=results9.keys())

# Printing the resulting DataFrame
print("\nCorrected Query 9 Results:")
print(df9)
```

# 10.Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for properties listed for sale in the year 2020. Exclude locations with fewer than 10 properties listed for sale in 2020 from the results.

```python
# Query 10: Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for properties listed for sale in the year 2020. Exclude locations with fewer than 10 properties listed for sale in 2020 from the results.
query10 = """
SELECT PD.l1, PD.l2, PD.l3, AVG(PPD.price / PD.surface_total) AS avg_price_per_sqm
FROM Property_Details PD
JOIN Property_Price_Details PPD ON PD.id = PPD.id
WHERE strftime('%Y', PD.created_on) = '2020'
GROUP BY PD.l1, PD.l2, PD.l3
HAVING COUNT(*) >= 10
ORDER BY avg_price_per_sqm DESC
LIMIT 3;
"""

# Executing the corrected query and retrieve the results into a DataFrame
with engine.connect() as connection:
    results10 = connection.execute(text(query10))
    df10 = pd.DataFrame(results10.fetchall(), columns=results10.keys())

# Printing the resulting DataFrame
print("\nCorrected Query 10 Results:")
print(df10)
```
